# 6.3 Graphing Infant Mortality on a map

```python
import altair as alt
from vega_datasets import data
counties = alt.topo_feature(data.us_10m.url, 'counties')

# Does this Json somehow have the map in it?
print(data.us_10m.url)
print(counties)

alt.Chart(counties).mark_geoshape().project(
    type='albersUsa').properties(
    width=500,
    height=300
)
```

```python
import numpy as np
import pandas as pd
```

```python
unemp_data = pd.read_csv('http://vega.github.io/vega-datasets/data/unemployment.tsv',sep='\t')
unemp_data.head()
```

Using the transform_lookup method we can arrange for the id in the geographic data to be matched against the id in our unemp_data data frame. 

```python
e=alt.Chart(counties).mark_geoshape(
).encode(
    color='rate:Q'
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(unemp_data, 'id', ['rate'])
).project(
    type='albersUsa'
).properties(
    width=500,
    height=300,
    title='Unemployment by County'
)

# can we put a tooltip on this?

e.save('unemployment.html')
e.display()
```

we need to write a function that takes a three letter country code as a parameter and returns the numeric code we lookup as an integer, lets call it get_num_code.

```python
# This function returns the numeric code from a countrycode
def get_num_code(countrycode):
    url='https://restcountries.eu/rest/v2/alpha/'
    return int(requests.get(url+countrycode).json()['numericCode'])
```

Once you write this function you can use as shown below:

NOTE: The copy of world_countries.csv already had CodeNum, but we get the idea. If we only had Code, we could create a new column with CodeNum easily as explained.

```python
wd.head(2)
```

```python
import time  #Library to record the current time
tick=time.time() # The current time before the creation of wd['id']
wd['id'] = wd['Code'].map(get_num_code) # This creates a column of id's for each country in wd
tock=time.time() # The current time after the creation of wd['id']
print("This computation was ",tock-tick," seconds long")
wd.head()
```

You can make a gray map of the world like this:

```python
countries = alt.topo_feature(data.world_110m.url, 'countries')

alt.Chart(countries).mark_geoshape(
    fill='#666666',
    stroke='white'
).properties(
    width=550,
    height=450
).project('equirectangular')
```

So, now you have the information you need to use the example of the counties above and apply that to the world below.

```python
base = alt.Chart(countries).mark_geoshape(
    fill='#666666',
    stroke='white'
).encode( 
    color='Infant mortality:Q'

).transform_lookup( 
    lookup='id',
    from_=alt.LookupData(wd, 'id', ['Infant mortality'])

).properties(
    width=550,
    height=450
).project('equirectangular')

base
```
