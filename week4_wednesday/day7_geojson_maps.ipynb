### Making Maps from GeoJSON

```python
import json
import requests
```

```python
import altair as alt
alt.renderers.enable('notebook')
alt.data_transformers.enable('json')
```

```python
#url = 'https://global.mapit.mysociety.org/area/239439.geojson'
url = 'https://cugir.library.cornell.edu/download/file/cugir-007961-geojson.json'

response = requests.get(url)
map_json = response.text

map_data = json.loads(map_json)
```

```python
alt.Chart(alt.Data(values= map_json)).mark_geoshape(stroke = 'black').project('albersUsa')
```
